```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats

print("pandas version:", pd.__version__)
print('numpy version:',  np.__version__)
print('matplotlib version:',  matplotlib.__version__)
```

```python
enzyme_base_dis = pd.read_csv('Enzyme_base_dis.csv',index_col=0)
enzyme_base_dis.head()
```

```python
fig,ax = plt.subplots(1,1,figsize=(18,3))

ax.tick_params(axis='both',labelsize=14,bottom=True,labelbottom=True)

ax.set_xlabel('Day',size=14)
ax.set_ylabel('Enzyme',size=14)
ax.axvline(x=365*3, color='k',linestyle='-',lw=3)
ax.axvline(x=365*6, color='k',linestyle='-',lw=3)

for replicate in range(enzyme_base_dis.shape[1]):
    ax.plot(enzyme_base_dis.iloc[:,replicate],color='grey')
ax.plot(enzyme_base_dis.mean(axis=1),color='red')

plt.show()
```

```python
enzyme_basex10_dis = pd.read_csv('Enzyme_basex10_dis.csv',index_col=0)
enzyme_basex10_dis.head()
```

```python
fig,ax = plt.subplots(1,1,figsize=(18,3))

ax.tick_params(axis='both',labelsize=14,bottom=True,labelbottom=True)

ax.set_xlabel('Day',size=14)
ax.set_ylabel('Enzyme',size=14)
ax.axvline(x=365*3, color='k',linestyle='-',lw=3)
ax.axvline(x=365*6, color='k',linestyle='-',lw=3)

for replicate in range(enzyme_basex10_dis.shape[1]):
    ax.plot(enzyme_basex10_dis.iloc[:,replicate],color='grey')
ax.plot(enzyme_basex10_dis.mean(axis=1),color='red')

plt.show()
```

```python
index_3   = [i for i in range(365*2+1,365*3+1)]
index_6   = [i for i in range(365*5+1,365*6+1)]
index_9   = [i for i in range(365*8+1,365*9+1)]
index_369 = index_3 + index_6 + index_9

enzyme_base_dis369    = enzyme_base_dis.iloc[index_369,:].reset_index(drop=True)

enzyme_basex10_dis369 = enzyme_basex10_dis.iloc[index_369,:].reset_index(drop=True)
```

```python
def confi_interval(df):
    alpha = 0.10
    
    lower = df.apply(
        lambda df: df.mean() - (stats.t.ppf(1 - alpha/2, df.count()-1) * df.std(ddof=1)/np.sqrt(df.count())),axis=1
    )
    upper = df.apply(lambda df: df.mean() + (stats.t.ppf(1 - alpha/2, df.count()-1) * df.std(ddof=1)/np.sqrt(df.count())),axis=1)
    
    return lower,upper
```

```python
bound369    = confi_interval(enzyme_base_dis369.iloc[:,:])

boundx10369 = confi_interval(enzyme_basex10_dis369.iloc[:,:])
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,3))
ax.axvline(x=365*1, color='grey',linestyle='--',lw=3)
ax.axvline(x=365*2, color='grey',linestyle='--',lw=3)
ax.set_xlim(-5,365*3+5)
#ax.set_ylim(0.8,1.0)

ax.tick_params(bottom=False,labelbottom=False)    
#ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)

#ax.set_xlabel('Year',size=16)
ax.set_ylabel("Drought tolerance",size=16,weight='normal', color='k')
plt.figtext(0.23,0.05,'Year-3',fontsize=14)
plt.figtext(0.50,0.05,'Year-6',fontsize=14)
plt.figtext(0.75,0.05,'Year-9',fontsize=14)

plt.figtext(0.23,0.9,'ambient',fontsize=14)
plt.figtext(0.50,0.9,'drought',fontsize=14)
plt.figtext(0.75,0.9,'ambient',fontsize=14)

ax.plot(enzyme_basex10_dis369.mean(axis=1)[365*1:365*3],c='k',linestyle='--')
ax.fill_between(x=enzyme_basex10_dis369.index[365*1:365*2],y1=boundx10369[0][365*1:365*2],y2=boundx10369[1][365*1:365*2], facecolor='red',alpha=1.0)
ax.fill_between(x=enzyme_basex10_dis369.index[365*2:365*3],y1=boundx10369[0][365*2:365*3],y2=boundx10369[1][365*2:365*3], facecolor='red',alpha=1.0)

ax.plot(enzyme_base_dis369.mean(axis=1),c='k',linestyle='--')
ax.fill_between(x=enzyme_base_dis369.index[365*0:365*1], y1=bound369[0][365*0:365*1],y2=bound369[1][365*0:365*1],facecolor='blue', alpha=1.0) 
ax.fill_between(x=enzyme_base_dis369.index[365*1:365*2], y1=bound369[0][365*1:365*2],y2=bound369[1][365*1:365*2],facecolor='blue', alpha=0.5) 
ax.fill_between(x=enzyme_base_dis369.index[365*2:365*3], y1=bound369[0][365*2:365*3],y2=bound369[1][365*2:365*3],facecolor='blue', alpha=0.5) 

#ax.legend()

#plt.savefig('legacy_sub.pdf',bbox_inches='tight')

plt.show()
```
