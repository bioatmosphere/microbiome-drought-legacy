# Taxon level

## Day 731

```python
day_732 = pd.concat([enzyme_base.loc[:,'732'],osmolyte_base.loc[:,'732'],yield_base.loc[:,'732']],axis=1,sort=False)
day_732.columns = ['Enzyme','Osmolyte', 'Yield']
day_732.head()
```

```python
total = day_732.sum(axis=1)
day_732_alive = 100*day_732[total>0].divide(total[total>0],axis=0)
day_732_alive.head()
```

## Define function getting data of a specific day

```python
def day_data(day):
    
    day = pd.concat([enzyme_base.loc[:,str(day)], osmolyte_base.loc[:,str(day)],yield_base.loc[:,str(day)]],axis=1,sort=False)
    day.columns = ['Enzyme','Osmolyte', 'Yield']
    total = day.sum(axis=1)
    day_alive = 100*day[total>0].divide(total[total>0],axis=0)
    
    return day_alive
```

## day 880

```python
day_880_alive = day_data(880)
day_1080_alive = day_data(1080)

day_1080_alive.head()
```

```python
# Scatter Plot
scale = 100
figure, tax = ternary.figure(scale=scale)
figure.set_size_inches(10, 10)

tax.scatter(day_732_alive.values, marker='.', color='red', label="Red Squares")
#tax.scatter(day_880_alive.values, marker='.', color='blue',label="Red Squares")
#tax.scatter(day_1080_alive.values, marker='.', color='orange',label="Red Squares")

tax.legend()

#tax.set_title("Scatter Plot", fontsize=20)
fontsize =12
tax.left_axis_label("Enzyme", fontsize=fontsize, offset=0.14)
tax.right_axis_label("Yield", fontsize=fontsize, offset=0.14)
tax.bottom_axis_label("Osmolyte", fontsize=fontsize, offset=0.14)

tax.gridlines(multiple=scale/10, color="blue", linewidth=0.5)
tax.boundary(linewidth=2.0)
tax.ticks(axis='lbr', linewidth=1, multiple=scale/10)
tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')

#tax.savefig('scatter_ternary.pdf')
tax.show()
```

# Community level

```python
community = pd.concat([enzyme_base.sum(axis=0),osmolyte_base.sum(axis=0),yield_base.sum(axis=0)],axis=1,sort=False)
community.columns = ['Enzyme','Osmolyte', 'Yield']
total = community.sum(axis=1)
community_alive = 100*community[total>0].divide(total[total>0],axis=0)
community_alive
```

```python
# Scatter Plot
scale = 100
figure, tax = ternary.figure(scale=scale)
figure.set_size_inches(10, 10)

tax.scatter(community_alive.values, marker='.', color='red', label="Red Squares")


tax.legend()

#tax.set_title("Scatter Plot", fontsize=20)
fontsize =12
tax.left_axis_label("Enzyme", fontsize=fontsize, offset=0.14)
tax.right_axis_label("Yield", fontsize=fontsize, offset=0.14)
tax.bottom_axis_label("Osmolyte", fontsize=fontsize, offset=0.14)

tax.gridlines(multiple=scale/10, color="blue", linewidth=0.5)
tax.boundary(linewidth=1.0)
tax.ticks(axis='lbr', linewidth=1, multiple=scale/10)
tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')

#tax.savefig('scatter_ternary.pdf')
tax.show()
```
