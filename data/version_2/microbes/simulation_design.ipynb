```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

print("pandas version:", pd.__version__)
print('numpy version:',  np.__version__)
print('matplotlib version:',  matplotlib.__version__)
```

```python
mic_base = pd.read_csv('Mic_base_20201.csv',index_col=0)
mic_base = mic_base/10000
mic_base.head()
```

```python
mic_base_dis = pd.read_csv('Mic_base_dis_20201.csv',index_col=0)
mic_base_dis = mic_base_dis/10000
mic_base_dis.head()
```

```python
base = pd.read_csv('forcing/base_forcing.csv', index_col=0)
base.head()
```

```python
basex4 = pd.read_csv('forcing/basex4_forcing.csv', index_col=0)
basex4.head()
```

```python
basex10 = pd.read_csv('forcing/basex10_forcing.csv', index_col=0)
basex10.head()
```

```python
year = 10
fig, (ax1,ax2) = plt.subplots(2, 1,figsize=(18,5))

ax1.plot(basex4['Psi'],color='red',linewidth=2,alpha=0.6)
ax1.plot(base['Psi'],color='blue',linewidth=2)

ax1.set_ylabel('Water potenial',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False,labelsize=14)    

for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

ax2.plot(basex4['Temp'],color='red',linewidth=2,alpha=0.6)    
ax2.plot(base['Temp'],color='blue',linewidth=2)

ax2.set_ylabel('Temperature',size=14)
ax2.set_xlim(-10,365*year+10)
ax2.tick_params(bottom=False,labelbottom=False, labelsize=14)    

for i in range(year-1):
    ax2.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)
    
plt.subplots_adjust(hspace=0.3)

plt.show()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize=(18,10))

ax1.plot(basex10['Psi'],color='red',linewidth=2,alpha=1.0)
ax1.plot(basex4['Psi'],color='orange',linewidth=2,alpha=1.0)
ax1.plot(base['Psi'],color='blue',linewidth=2)

ax1.set_ylabel('$\Psi$(MPa)',size=14)
ax1.set_xlim(-10,365*year+10)
ax1.tick_params(bottom=False,labelbottom=False)
ax1.yaxis.set_tick_params(labelsize=14)
ax1.text(0-100,6,'A',fontsize=14, weight='heavy')
for i in range(year-1):
    ax1.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)

#legend
blue_patch   = mpatches.Patch(color='blue',label='ambient')
orange_patch = mpatches.Patch(color='orange',label='moderate')
red_patch    = mpatches.Patch(color='red',label='severe')
ax1.legend(handles=[blue_patch,orange_patch,red_patch],fontsize=12,frameon=False)


#ax2.plot(basex4['Temp'],color='red',linewidth=2,alpha=0.6)    
ax2.plot(base['Temp'],color='blue',linewidth=2)

ax2.set_ylabel('Temperature(\u00B0C)',size=14)
ax2.set_xlim(-10,365*year+10)
ax2.tick_params(bottom=False,labelbottom=False)
ax2.yaxis.set_tick_params(labelsize=14)
ax2.text(0-100,33,'B',fontsize=14, weight='heavy')

for i in range(year-1):
    ax2.axvline(x=365*(i+1)-1, color='k',linestyle=':',lw=2)


ax3.set_xlim(-10,365*10+10)
ax3.set_ylim(-1,35)
ax3.set_ylabel('Microbes\n(mgC cm$^-$$^3$)',size=14)
ax3.set_title('Default Mode',size=14)
ax3.tick_params(bottom=False,labelbottom=False)    
ax3.yaxis.set_tick_params(labelsize=14)
ax3.axvline(x=365*3,color='grey',linestyle=':',lw=2)
ax3.axvline(x=365*6,color='grey',linestyle=':',lw=2)
ax3.text(0-100,37,'C',fontsize=14, weight='heavy')

ax3.text(365*1,30,'ambient',fontsize=14, weight='heavy')
ax3.text(365*4,30,'drought',fontsize=14, weight='heavy')
ax3.text(365*7,30,'ambient',fontsize=14, weight='heavy')

ax3.stackplot(range(1,365*10+1),mic_base.iloc[:,1:])



ax4.set_xlim(-10,365*10+10)
ax4.set_ylim(-1,35)
ax4.set_ylabel('Microbes\n(mgC cm$^-$$^3$)',size=14)
ax4.set_title('Dispersal Mode',size=14)
ax4.tick_params(bottom=False,labelbottom=False)    
ax4.yaxis.set_tick_params(labelsize=14)
ax4.axvline(x=365*3,color='grey',linestyle=':',lw=2)
ax4.axvline(x=365*6,color='grey',linestyle=':',lw=2)
ax4.text(0-100,36,'D',fontsize=14, weight='heavy')
ax4.stackplot(range(1,365*10+1),mic_base_dis.iloc[:,1:])

plt.subplots_adjust(hspace=0.2)

plt.show()
```
