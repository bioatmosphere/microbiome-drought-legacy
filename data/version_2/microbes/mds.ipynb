## generalization

```r
data <- function(data_csv){
    rownames = data_csv[,1]
    colnames = paste("D",1:1095,sep="")
    data_csv = data_csv[,2:1096]
    rownames(data_csv) = rownames
    colnames(data_csv) = colnames
    t(data_csv)  
}
```

```r
mic_base369 <- read.csv('mic_base369.csv')
mic_basex4369 <- read.csv('mic_basex4369.csv')
mic_basex10369 <- read.csv('mic_basex10369.csv')

mic_base_dis369 <- read.csv('mic_base_dis369.csv')
mic_basex10_dis369 <- read.csv('mic_basex10_dis369.csv')
```

```r
mic_base369    <- data(mic_base369)
mic_basex4369  <- data(mic_basex4369)
mic_basex10369 <- data(mic_basex10369)

mic_base_dis369 <- data(mic_base_dis369)
mic_basex10_dis369 <- data(mic_basex10_dis369)
```

```r
dim(mic_base369)
```

```r
sub_mic_base369 <- mic_base369[(365*2+1):(365*3),]
sub_mic_basex4369 <- mic_basex4369[(365*2+1):(365*3),]
sub_mic_basex10369 <- mic_basex10369[(365*2+1):(365*3),]
```

```r
NMDS_base    <- metaMDS(sub_mic_base369,k=2)
NMDS_basex4  <- metaMDS(sub_mic_basex4369,k=2)
NMDS_basex10 <- metaMDS(sub_mic_basex10369,k=2)
```

```r
ordiplot(NMDS_base,type="n")
#orditorp(example_NMDS,display="species",col="red",air=0.01)
orditorp(NMDS_base,display="sites",col="blue",cex=1.00,air=0.01)
```

```r
ordiplot(NMDS_basex4,type="n")
#orditorp(example_NMDS,display="species",col="red",air=0.01)
orditorp(NMDS_basex4,display="sites",col="blue",cex=1.00,air=0.01)
```

```r
ordiplot(NMDS_basex10,type="n")
#orditorp(example_NMDS,display="species",col="red",air=0.01)
orditorp(NMDS_basex10,display="sites",col="blue",cex=1.00,air=0.01)
```

## year-9

```r
group <- rbind(mic_base369[(365*2+1):(365*3),],mic_basex4369[(365*2+1):(365*3),],mic_basex10369[(365*2+1):(365*3),])
rownames(group) = c(rep('base',365),rep('basex4',365),rep('basex10',365))
group
```

## dispersal

```r
group_dis <- rbind(mic_base_dis369[(365*2+1):(365*3),],mic_basex10_dis369[(365*2+1):(365*3),])
rownames(group_dis) = c(rep('base',365),rep('basex10',365))
head(group_dis)
```

```r
NMDS_group_dis <- metaMDS(group_dis,k=2,trymax=100)
```

```r
ordiplot(NMDS_group_dis,type="n")
#orditorp(example_NMDS,display="species",col="red",air=0.01)
orditorp(NMDS_group_dis,display="sites",col=c(rep('blue',365),rep('orange',365),rep('red',365)),cex=1.00,air=0.01)
```
